Working space to **prioritize**, **scope**, and **pair** predictive vs explanatory models before building notebooks under `ml-pipelines/`.

See `pipelineGuidelines.md` and `IntexContext.txt` (IS455) for required sections and deployment expectations.

## Rating scale (gut check)

| Score | Meaning |
| :---: | --- |
| **5/5** | Strong fit for course rubric + clear story + good data join |
| **4/5** | High value; may need careful leakage / ethics / scope |
| **3/5** | Worth doing if time; may be explanatory-heavy or noisier |
| **2/5** | Nice supplement or narrative bridge; not a first pipeline |

## Candidate pipelines

---

### 1. Expected next donation amount or total giving — **3/5**

- **Question:** Who might give more if asked (upgrade potential)?
- **Data:** Donation history by `supporter_id`; campaign flags; recurring; donation type mix.
- **Predictive:** Regression on future amount or bucket.
- **Explanatory:** Factors correlated with larger gifts (causality: careful).

---

### 2. Campaign effectiveness — **3.5/5**

- **Question:** Which campaigns actually lift donation value vs noise?
- **Data:** `campaign_name`, `donation_date`, `channel_source`, amounts; optional join to allocations by program area.
- **Explanatory:** Campaign periods vs baseline; difference-in-means or regression with time controls.
- **Predictive (optional):** Forecast donations next month by campaign calendar.

---

### 3. Safehouse strain / capacity pressure — **4/5**

- **Question:** Which safehouses show rising incidents or falling average progress?
- **Data:** `safehouse_monthly_metrics`, `incident_reports`, resident counts.
- **Predictive:** Forecast next month’s incident load or stress index.
- **Explanatory:** Correlates of high-incident months.
- **Notebook started:** `safehouse-strain-forecast.ipynb`

---

### 4. Intervention / program effectiveness — **3/5**

- **Question:** Which intervention categories associate with better education or health trajectories?
- **Data:** `intervention_plans` joined to `education_records` / `health_wellbeing_records` by resident.
- **Explanatory:** Simple panel or resident-level patterns; state **non-causal** unless design supports more.

---

### 5. Risk / “falling through the cracks” flag — **5/5**

- **Question:** Which residents look like they’re struggling (incidents, visit outcomes, stalled progress)?
- **Data:** Rolling signals from `incident_reports`, `process_recordings` (e.g. concern flags), `home_visitations`, trends in education/health.
- **Predictive:** Binary or ordinal “needs review soon.”
- **Explanatory:** Operational factors associated with worse trajectories (privacy + minors: careful in writeup).

---

### 6. Reintegration readiness — **4/5**

- **Question:** Who looks closer to successful transition (aligned with `reintegration_status`, visit types, plans)?
- **Data:** `residents`, `intervention_plans`, `home_visitations` (e.g. reintegration assessment); prefer **aggregates** over raw narrative text if scoping.
- **Predictive:** Status or time-to-milestone.
- **Explanatory:** Drivers of `reintegration_status` or progress.

---

### 7. Post → donation linkage — **5/5**

- **Question:** What post features associate with referred donations (`referral_post_id`, engagement, topic, time of day)?
- **Data:** `social_media_posts` + `donations` where social referral exists.
- **Predictive:** Referral or value from post attributes.
- **Explanatory:** Regression with **selection bias** discussion (boosted posts, campaigns).

---

### 8. Engagement vs “vanity” — **2/5**

- **Question:** What distinguishes high-engagement posts from posts that drive clicks/donations?
- **Data:** Engagement vs `click_throughs`, `donation_referrals` (if populated consistently).
- **Explanatory:** Compare outcomes by `post_type`, `content_topic`, `platform`.

---

### 9. Bridge: outreach → money → outcomes — **2/5**

- **Question:** Do months with stronger social-referred giving align with better aggregates (privacy-safe only)?
- **Data:** Time-series join of donation totals with `safehouse_monthly_metrics` or `public_impact_snapshots`.
- **Caution:** Ecological fallacy; keep claims modest; strong **narrative** for Hiraya if executed carefully.

---

## Suggested priority order (by rating)

1. **5/5** — Risk / “falling through the cracks”; Post → donation linkage  
2. **4/5** — Safehouse strain (in progress); Reintegration readiness  
3. **3.5/5** — Campaign effectiveness  
4. **3/5** — Next donation / upgrade; Intervention effectiveness  
5. **2/5** — Engagement vs vanity; Outreach → money → outcomes bridge
