# Social content drivers (Hiraya Haven)

**Business question (*IntexContext*):** What kinds of posts appear to lead to donations versus likes-only vanity metrics?

- **Explanatory focus:** Roll up `social_media_posts` by topic, media type, post type, and CTA; inspect referral rate, total estimated PHP, and boosted share.
- **Predictive extension (offline):** Classify high-referral posts using interpretable logistic regression and/or shallow tree; validate with time-based split.
