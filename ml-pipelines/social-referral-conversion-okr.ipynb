# OKR — Social referral conversion (quarterly)

**Metric:** Share of posts in a calendar quarter with `donation_referrals >= 1`.

**Why it matters:** Directly ties the outreach domain to fundraising effectiveness (IS413/455 alignment).

**Targets:** Stored in `okr_targets` with `metric_key = outreach.social.referral_conversion_rate`.

## Deployment
`GET /api/okrs/outreach/social/referral-conversion/quarterly` · Admin `PUT` on same resource namespace · Pipeline id: `outreach-social-referral-okr`.
