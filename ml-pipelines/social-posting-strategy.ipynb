# Social posting strategy (Hiraya Haven)

**Business question (*IntexContext*):** When should we post to maximize donation-linked outcomes given limited marketing staff?

- **Explanatory focus:** Compare hour-of-day and weekday buckets on donation referrals, estimated attributed PHP, engagement, and CTR. Associations only; confounders include paid boosts and campaign calendars.
- **Predictive extension (offline):** Train a simple model to predict `donation_referrals > 0` from hour, weekday, platform, topic—report ROC-AUC / calibration for prioritizing test slots.
