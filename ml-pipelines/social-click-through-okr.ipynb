# OKR — Social click-through rate (quarterly)

**Metric:** Aggregate CTR per quarter: `sum(click_throughs) / sum(impressions)` across posts dated in the quarter.

**Why it matters:** Measures reach-to-action efficiency on the primary donor acquisition channel before gifts are attributed.

## Deployment
`GET /api/okrs/outreach/social/click-through/quarterly` · Pipeline id: `outreach-social-ctr-okr`.
